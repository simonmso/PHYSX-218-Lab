## Part 1 - Unknown resistor

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import error_rules as er

amps = [0.0204, 0.041, 0.0614, 0.0467, 0.0798]
unc_amp = 0.00005
volts = [1.3, 3.13, 4.69, 4.58, 6.09]
unc_volt = 0.02
Rs  = []
unc_Rs = []

for i in range(len(amps)):
    Rs.append(volts[i] / amps[i])
    unc_Rs.append(er.rule_4(Q=Rs[i], values=[volts[i], amps[i]], uncertainties=[unc_volt, unc_amp], exponents=[1, -1]))

print('Resistance:')
print(Rs)
print('Errors:')
print(unc_Rs)
#-------------------------------------------#
#Data Section - Create Arrays for data. Perform necessary calculations
#CHANGE THE VARIABLE NAMES and numbers to match your data
xvariable_changeme = np.array(amps) #what are units?
yvariable_changeme = np.array(volts) #what are units?


#--------------------------------------------#
#Create arrays for uncertainties
#CHANGE THE VARIABLE NAME and numbers to match your data 
err_yvariable_changeme = np.array([unc_volt] * len(volts))


#--------------------------------------------#
#Re-assign variables as x, y, dy so that the following code may remain generic

x = xvariable_changeme   #this should be the array you want to plot on the x axis
y = yvariable_changeme
dy = err_yvariable_changeme  #this should be your error in y array

#----------------------------------------------#
#Don't need to change anything in this section!
 
#Find the intercept and slope, b and m, from Python's polynomial fitting function
b,m=np.polynomial.polynomial.polyfit(x,y,1,w=dy)

#Write the equation for the best fit line based on the slope and intercept
fit = b+m*x

#Calculate the error in slope and intercept 
#def Delta(x, dy) is a function, and we will learn how to write our own at a later date. They are very useful!
def Delta(x, dy):
    D = (sum(1/dy**2))*(sum(x**2/dy**2))-(sum(x/dy**2))**2
    return D
 
D=Delta(x, dy)
 
dm = np.sqrt(1/D*sum(1/dy**2)) #error in slope
db = np.sqrt(1/D*sum(x**2/dy**2)) #error in intercept

#Calculate the "goodness of fit" from the linear least squares fitting document
def LLSFD2(x,y,dy):
    N = sum(((y-b-m*x)/dy)**2)
    return N
                      
N = LLSFD2(x,y,dy)

#-----------------------------------------------------------------------#
#Plot data on graph. Plot error bars and place values for slope, error in slope
#and goodness of fit on the plot using "annotate"

plt.figure(figsize=(15,10))
 
plt.plot(x, fit, color='green', linestyle='--')
plt.scatter(x, y, color='blue', marker='o')
 
#create labels  YOU NEED TO CHANGE THESE!!!
plt.xlabel('Current (Amps)')
plt.ylabel('Voltage (Volts)')
plt.title('Resistance')
 
plt.errorbar(x, y, yerr=dy, xerr=None, fmt="none") #don't need to plot x error bars
 
plt.annotate('Slope (Ohms) = {value:.{digits}E}'.format(value=m, digits=2),
             (0.05, 0.9), xycoords='axes fraction')
 
plt.annotate('Error in Slope (Ohms) = {value:.{digits}E}'.format(value=dm, digits=1),
             (0.05, 0.85), xycoords='axes fraction')
 
plt.annotate('Goodness of fit = {value:.{digits}E}'.format(value=N, digits=2),
             (0.05, 0.80), xycoords='axes fraction')

plt.show()
```

## Part 2 - Series Circuit

```python
volts = [2.6, 0.5]
v_tot = 3.14
amps = [0.026, 0.0261]
Rs = []
unc_Rs = []

for i in range(len(volts)):
    Rs.append(volts[i] / amps[i])
    unc_Rs.append(er.rule_4(Q=Rs[i], values=[volts[i], amps[i]], uncertainties=[unc_volt, unc_amp], exponents=[1, -1]))

R_tot_exp = sum(Rs)
unc_R_tot = er.rule_3(unc_Rs)
print(f"Total resistance: {R_tot_exp:.1f} ± {unc_R_tot:.3f} Ohms")
```

## Part 3 - Parallel Circuit

```python
volts = [3.01, 2.97]
amps = [0.0301, 0.142]
Rs = []
unc_Rs = []

for i in range(len(amps)):
    Rs.append(volts[i] / amps[i])
    unc_Rs.append(er.rule_4(Q=Rs[i], values=[volts[i], amps[i]], uncertainties=[unc_volt, unc_amp], exponents=[1, -1]))

exp_R_tot = 1 / ((1 / Rs[0]) + (1 / Rs[1]))
unc_R_tot = er.rule_3(unc_Rs)

print(f"Total resistance: {exp_R_tot:.1f} ± {unc_R_tot:.3f} Ohms")
```
